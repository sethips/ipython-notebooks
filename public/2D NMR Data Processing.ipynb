# 2D NMR Data Processing

This script processes 2D NMR data (in Pathomx data file standard format) and plots the various classes.


*Note: Load the datafile from the filename specified below. Note that if you run this script from within QtIPy on a watch-folder trigger (with iterate files) it will automagically run against any file you drop into that folder.*

```python
import os
import pandas as pd
import numpy as np

if 'qtipy' in vars(): # We're in a auto-session in QtIPy
    filename = os.path.join(qtipy['watched_folder'], qtipy['filename'])
    
else: # Non-automated - edit this filename to try out on your own data
    filename = "/Users/mxf793/Data/THPNH/Extract/thp1_2d_jres_bml_nmr.csv"
```

```python
data = pd.read_csv(filename, dtype={'Sample': str, 'Class': str}, index_col=['Class', 'Sample'] )
```

```python
classes = set([k for k,v in data.index.values])
classes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from collections import OrderedDict
from mplstyler import StylesManager

styles = StylesManager()

plt.figure()

x = range(data.shape[1])
plots=OrderedDict()
for c in classes:
    y = np.mean(data.loc[c].values, axis=0)
    s = styles.get_style_for_class(c)
    plots[c], =plt.plot(x, y, **s.line_kwargs)
    
legend = plt.legend(list(plots.values()),
   list(plots.keys()),
   loc='best') #, bbox_to_anchor=(1, 1))
legend.get_frame().set_facecolor('k')                      
legend.get_frame().set_alpha(0.05)     
    
if 'qtipy' in vars():
    plt.savefig(qtipy['output_path'] + 'plot.png', dpi=200) # Save the figure 
    os.rename(filename, qtipy['output_path'] + qtipy['filename'] ) # Move the source data into the same folder
```
