```python
# Plot for metabolite metabolite_plot
metabolite_plot = "Lactate"
```

```python
import pandas as pd
```

```python
d = pd.io.parsers.read_csv('/Users/mxf793/Data/MoMetab+L#1/ELISA.csv')
d
```

```python
single = d[ pd.Series([ (metabolite_plot in s or 'RPMI' in s) and '0.001' in s for s in d['Class']])]

single
```

```python
single['Mo'] = pd.Series([ s[:2] for s in single['Class']], index=single.index, dtype=str)
single['Condition'] = pd.Series([ 'RPMI' if 'RPMI' in s else metabolite_plot for s in single['Class']], index=single.index, dtype=str)
single
```

```python
del single['Sample']
del single['Class']
```

```python
#data = lactate.set_index(['Mo','Condition'])

group = single.groupby(['Mo','Condition'])
m = group.mean()
e = group.std()

data = single

```

```python
from matplotlib import pyplot as plt
from matplotlib import rcParams
rcParams['font.serif'] = ['Computer Modern Roman', 'Times New Roman']
rcParams['font.family'] = 'sans-serif'
```

```python
ax = m.plot(kind='bar', title='Il-6 production by M1 and M2 with 0.001mM %s' % metabolite_plot, yerr=e.values, color='k', ecolor='k')
ax.set_ylabel('Il-6 pg/ml')
ax.set_xlabel('Monocyte subset + Culture conditions')
ax.set_xticklabels( ('M1 + %s' % metabolite_plot, 'M1 + RPMI', 'M2 + %s' % metabolite_plot, 'M2 + RPMI'), rotation=0 )
ax.get_legend().remove()
ax.grid(False)

fig = ax.get_figure()
fig.set_size_inches((6,6))
fig.savefig('/Users/mxf793/Data/MoMetab+L#1/m1m2-%s-il6.png' % metabolite_plot, dpi=200)
```

```python
from scipy.stats import ttest_ind
```

```python
m1r = data[(data["Mo"]=="M1") & (data["Condition"]=="RPMI") ]["IL-6"]
m1l = data[(data["Mo"]=="M1") & (data["Condition"]==metabolite_plot) ]["IL-6"]

r = ttest_ind(m1r.values, m1l.values)
print("M1 RPMI vs %s t=%.2f, p=%.2f" % (metabolite_plot, r[0], r[1] ) )

m2r = data[(data["Mo"]=="M2") & (data["Condition"]=="RPMI") ]["IL-6"]
m2l = data[(data["Mo"]=="M2") & (data["Condition"]==metabolite_plot) ]["IL-6"]

r = ttest_ind(m2r.values, m2l.values)
print("M2 RPMI vs %s t=%.2f, p=%.2f" % (metabolite_plot, r[0], r[1] ) )
```
