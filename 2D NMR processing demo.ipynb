# Setup

Load the datafile from the filename specified above. Note that if you run this script from within QtIPy on a watch-folder trigger (with iterate files) it will automagically run against any file you drop into that folder.

```python
data = pd.read_csv(filename, dtype={'Sample': str, 'Class': str}, index_col=['Class', 'Sample'] )
```

```python
classes = set([k for k,v in data.index.values])
classes
```

# First use

Now we've got the classes for the data plotted, we can now plot the mean values (of the spectra). To get a style assignment for a given class we need to call `.get_style_for_class()` on the styles manager. This returns a useable style object that can output the keyword arguments needed to correctly style matplotlib plots.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from collections import OrderedDict
from mplstyler import StylesManager

styles = StylesManager()

plt.figure()

x = range(data.shape[1])
plots=OrderedDict()
for c in classes:
    y = np.mean(data.loc[c].values, axis=0)
    s = styles.get_style_for_class(c)
    plots[c], =plt.plot(x, y, **s.line_kwargs)
    
legend = plt.legend(list(plots.values()),
   list(plots.keys()),
   loc='best') #, bbox_to_anchor=(1, 1))
legend.get_frame().set_facecolor('k')                      
legend.get_frame().set_alpha(0.05)     
    
if 'qtipy' in vars():
    plt.savefig(qtipy['output_path'] + 'plot.png', dpi=200) # Save the figure 
    os.rename(filename, qtipy['output_path'] + qtipy['filename'] ) # Move the source data into the same folder
```

# Second use

Now we've called `styles.get_style_for_class()` with a class (group) name the returned style is permanantly assigned for that class. Any subsequent calls with the same name will return the same class

```python
plt.figure()

for c in classes:
    y = np.mean(data.loc[c].values, axis=0)
    s = styles.get_style_for_class(c)
    plots[c], =plt.plot(x, y, **s.kwargs)
    
legend = plt.legend(list(plots.values()),
   list(plots.keys()),
   loc='best') #, bbox_to_anchor=(1, 1))
legend.get_frame().set_facecolor('k')                      
legend.get_frame().set_alpha(0.05)     
    
```
